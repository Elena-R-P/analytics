```python
import pyodide_http
pyodide_http.patch_all()
import warnings
warnings.filterwarnings('ignore')
import micropip
await micropip.install('pandas')
await micropip.install('openpyxl')
import pandas as pd

# file with data
excel_file = pd.ExcelFile("practice_forecasting_accuracy.xlsx", engine='openpyxl')

data = pd.read_excel(excel_file)
data.head()

# Calculate MSE
mse = ((data['Actual Projects'] - data['Forecasted Projects']) ** 2).mean()
# Print MSE
print(f'The Mean Squared Error (MSE) is: {mse}')

# Calculate MAE
mae = (data['Actual Projects'] - data['Forecasted Projects']).abs().mean()
# Print MAE
print(f'The Mean Absolute Error (MAE) is: {mae}')

# Calculate MAPE
mape = (abs(data['Actual Projects'] - data['Forecasted Projects'])/data['Actual Projects']).mean()*100
# Print MAPE
print(f'Mean Absolute Percentage Error (MAPE) is: {mape}')
```

A MAPE of 6.31% means the forecasts are, on average, off by 6.31% of the actual contracts, which is a relatively low error rate and indicates the forecasting model is performing well. This is particular useful communicating accuracy to stakeholders and guiding resource allocation.
