## Monte Carlo Simulation by building a single-instance model
+ Random variables represents uncertain numerical outcomes
+ Probability distribution describe how the values of a random variable are spread
+ Random sampling draws values from a probability distribution to simulate different scenarios

```python
import warnings
warnings.filterwarnings('ignore')
import pyodide_http
pyodide_http.patch_all()
import micropip
await micropip.install('numpy')
await micropip.install('matplotlib')
# import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Determine the probability distribution and generate random samples
np.random.seed(42) # Setting the random seed for reproducibility
# parameters
num_days = 30
daily_revenue = np.random.randint(200, 1001, num_days)

# Perform the simulation
average_daily_revenue = np.mean(daily_revenue)
monthly_revenue = average_daily_revenue * num_days
print(f'The estimated average monthly revenue is ${monthly_revenue:.2f}')

# Analyze the results
plt.hist(daily_revenue, bins=10, edgecolor='black')
plt.title('Distribution of Simulated Daily Revenue')
plt.xlabel('Revenue ($)')
plt.ylabel('Frequency')
plt.show()
```

1. Average Monthly Revenue<br/>
The estimated average monthly revenue from the simulation provides a baseline for financial planning. The company can use this estimate for budgeting, setting realistic sales targets, and making informed decisions about expenses and investments
2. Capturing Variability<br/>
Random sampling generates a range of possible daily revenue based on historical data. The Monte Carlo simulation helps the company capture the natural variability in daily revenue, reflecting real-world fluctuations and uncertainties
3. Visualizing Uncertainty<br/>
The company can expect fewer days with revenue close to the lower end of the range (250 to 400). This indicates that low-revenue days are less common. There will be more days where the revenue falls in the moderate to high range (500 to 1000). this suggests that the business is more likely to exoerience days with higher revenue, which is a positive sign for financial stability.
4. Scenario Testing<br/>
5. The company can explore the impact of different business strategies or market conditions on revenue, helping them make informed decisions and plan for potential risks.

## Full Monte Carlo Simulation

```python
import warnings
warnings.filterwarnings('ignore')
import pyodide_http
pyodide_http.patch_all()
import micropip
await micropip.install('numpy')
await micropip.install('matplotlib')
# import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Determine the probability distribution and generate random samples
np.random.seed(42) # Setting the random seed for reproducibility

# parameters
num_days = 30
num_simulations = 1000
min_revenue = 200
max_revenue = 1000

# Generate random samples for all simulations at once
daily_revenues = np.random.randint(min_revenue, max_revenue + 1, (num_simulations, num_days))

# Perform the simulation
monthly_revenues = np.sum(daily_revenues, axis=1)
average_monthly_revenue = np.mean(monthly_revenues)
median_monthly_revenue = np.median(monthly_revenues)
revenue_std_dev = np.std(monthly_revenues)

print(f'Estimated Average Monthly Revenue is ${average_monthly_revenue:.2f}')
print(f'Median Monthly Revenue is ${median_monthly_revenue:.2f}')
print(f'Standard Deviation of Monthly Revenue ${revenue_std_dev:.2f}')

# Analyze the results
plt.hist(monthly_revenues, bins=30, edgecolor='black')
plt.title('Distribution of Simulated Monthly Revenue')
plt.xlabel('Revenue ($)')
plt.ylabel('Frequency')
plt.axvline(average_monthly_revenue, color='r', linestyle='dashed', linewidth=1, label=f'Average:${average_monthly_revenue:.2f}')
plt.axvline(median_monthly_revenue, color='g', linestyle='dashed', linewidth=1, label=f'Median:${median_monthly_revenue:.2f}')
plt.legend()
plt.show()
```

The average monthly revenue (mean) is calculated by summing all the simulated monthly revenues and dividing by the number of simulations. The average monthly revenue provides an estimate of what the typical monthly revenue might be, considering the variability in daily revenues due to factors like promotions, holidays, and market trends. This figure is useful for budgeting, financial planning, and setting sales targets.<br/>
The median monthly revenue provides an estimate that is less affected by extreme values (outliers) compared to the average. This is particularly useful if there are occasional very high or very low daily revenues that could skew the average. The median gives more robust measure of the central tendency of the monthly revenues.<br/>
The standard deviation of 1,230.39 in the context of the company represents the variability or dispersion on the simulated monthly revenues. This value provides insight into how much the monthly revenues are expected to fluctuate around the average monthly revenue.
