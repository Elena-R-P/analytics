### Exploring Data Using Pandas

```python
file_path = 'https://raw.githubusercontent.com/sophiaAcademics/BDA_Excel/main/StudentData/Tutorials/Unit5/5.1.1/fruit.xlsx'

data = pd.read_excel(file_path)
#print(data) # print entire DataFrame
print(data.head()) # displays only first few rowsm practical for previewing the content
#print(data.info()) # provide a concise summary of a DataFrame
#print(data.describe()) # provide statistical summaries, such as mean,median,st dev
data.loc[2, "Product"] # to view a specific row ex: "Bananas"

# Output all Products
print("All fruits in the Product list:\n")
for x in range(len(data)):
    print(data.loc[x, "Product"])
```

### Visualizing Data with Matplotlib

```python
plt.scatter(data["Price (per lb)"], data["Sales (quanitity)"])
plt.xlabel("Price") # Label for the x-axis
plt.ylabel("Quantity") # Label for the y-axis
plt.title("Scatter Plot of Price vs Quantity")
plt.show()
```

```python
# Add a new column "Revenue"
data['Revenue'] = data['Sales (quanitity)'] * data['Price (per lb)']
print(data.head())
```

```python
plt.scatter(data["Price (per lb)"], data["Revenue"], color='purple', alpha=0.7)
plt.xlabel("Price (per lb)") # Label for the x-axis
plt.ylabel("Revenue") # Label for the y-axis
plt.title("Scatter Plot of Price vs Revenue")
plt.show()
```

```python
plt.scatter(data["Date"], data["Sales (quanitity)"])
plt.xlabel("Date") # Label for the x-axis
plt.ylabel("Quantity") # Label for the y-axis
plt.title("Scatter Plot of Date vs Quantity")
plt.show()
```
