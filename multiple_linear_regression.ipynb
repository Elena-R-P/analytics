```python
#product_launch.xlsx
# The goal is to create a multiple linear regression model
# that reliably generates new data.
import pandas as pd
import micropip
await micropip.install('pandas')
await micropip.install('openpyxl')

# excel file containing the product launch data
excel_file = pd.ExcelFile("product_launch.xlsx", engine='openpyxl')
# read the file into a pandas data frame
data = pd.read_excel(excel_file)

# Build a multiple linear regression model using k-fold cross-validation and 
# use RMSF as the assessment measure.
# Import the cross_val_score function for performing cross validation
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
import numpy as np

# select the features and target variable
features = ['Price', 'Marketing_Spend', 'Social_Media_Mentions', 'Pre_Orders', 'Customer_Reviews', 'Average_Rating']
target = 'Success_Score'
# prepare the future matrix (x) and target vector (y)
x = data[features]
y = data[target]

# initialize the model
model = LinearRegression()
# set a random seed for reproducibility
# shuffling process arranges the data in the same way every time
np.random.seed(42)
# preform 5-fold cross-validation
mse_scores = cross_val_score(model, x, y, cv=5, scoring='neg_mean_squared_error')
# calculate the RMSE across all folds
rmse_scores = (-mse_scores) ** 0.5
average_rmse = rmse_scores.mean()
# print the average RMSE
print(f"Average Root Mean Squared Error from 5-fold Cross-Validation: {average_rmse:.0f}")
```

###Classification and Regression Tree (CART)
+ Root Node - the starting point on the tree, represents the first decision you need to make
+ Internal Node - points where we make more decisiond based on different questions or tests
+ Branches - lines that connect the nodes, show possible answers to each question
+ Leaf Node - the end points of the tree, each node represents a final decision or outcome
